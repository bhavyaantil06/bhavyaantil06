SCATTER PLOT

```python
newx = np.random.randint(0,100,10)
newy = np.random.randint(0,100,10)
print(type(newy))
labels = ['A', 'B', 'C', 'D','E','F','G','H','I','J']

plt.scatter(newx, newy , marker= 'o', color='red',edgecolors='k', label="Data Points",s=100)  
# marker :  , o.o,+xXDdvV^
# colormarker color
# edgecolorsmarker edge color name
# smarker size
# labellegend string

for label, xi , yi in zip(labels,newx, newy):
    plt.text(xi+1, yi+1,label, va='bottom')

plt.xlabel('Random')
plt.ylabel('Random')
plt.title("Histogram")
```

Multiple plots

```python
#  making a sin graph and a cosine graph
# plt.subplot(211,figsize=(8, 6), sharex=True, gridspec_kw={'hspace': 0.5})
fig, axs = plt.subplots(2, 1, figsize=(8, 6), sharex=True, gridspec_kw={'hspace': 0.5});

x = np.linspace(0,2*np.pi,100)
# print(np.linspace(0,2*np.pi,100))
y = np.sin(x)
axs[0].plot(x,y,color='r');
axs[0].grid(True,color='k',linewidth='0.4')
axs[0].set_title('Sine graph')
axs[0].set_xlabel('Degree')
axs[0].set_ylabel('Amplitude')


plt.subplot(212)
# x = np.linspace
axs[1].set_xlabel('Degree')
axs[1].set_ylabel('Amplitude')
y = np.cos(x)
axs[1].plot(x, y, color="b");
axs[1].set_title('Cosine graph')
axs[1].grid(True,color='k')


'''plt.subplots(2, 1, ...) : This function creates a figure with 2 rows and 1 column of subplots. It means that the figure will have two subplots stacked vertically.

fig, axs = ... : This line unpacks the result of plt.subplots() into two variables: fig and axs.

fig is the Figure object representing the entire figure.
axs is a NumPy array or a list of Axes objects representing each subplot. In this case, since there are two subplots, axs will be an array or list with length 2, where each element represents a subplot.
figsize=(8, 6) : Specifies the size of the figure. Here, the figure will have a width of 8 inches and a height of 6 inches.

sharex=True : Shares the x-axis between the subplots. This means that both subplots will have the same range and ticks on the x-axis.

gridspec_kw={'hspace': 0.5} : This parameter is used to adjust the spacing between subplots. gridspec_kw is a dictionary of keyword arguments passed to plt.subplots() to adjust the layout. Here, {'hspace': 0.5} sets the horizontal spacing between subplots to be 0.5 times the height of the subplot. Increasing this value increases the vertical space between the subplots.''';
```
